## 대용량 저장장치

- 하드디스크
  - 자기적으로 기록하여 저장할 수 있는 플래터, 암 등으로 이루어진 디스크
  - 디스크는 대량의 보조 저장장치를 제공

- SSD (Solid State Disk)
  - 하드 디스크처럼 사용되는 비휘발성 메모리
  - 기계적 이동 부품이 없어 신뢰도↑ 
  - 탐색, 지연이 없어 속도↑
  - 전력소모↓

- 자기테이프
  - 초기의 보조 저장장치 매체. 현재는 백업용, 장기저장용으로 사용

## 스토리지 - 구성 방식으로 구분

- DAS(Direct Attached Storage)

  ![스크린샷 2022-12-01 오후 11 40 06](https://user-images.githubusercontent.com/96982072/205081096-ab4ee891-a591-4565-8168-5accfd3802e8.png)

  
  - 서버가 채널(SCSI 또는 광섬유)을 통해 저장장치에 직접 연결하는 방식
  - 확장이 쉬움, 속도가 빠름
  - 연결 수에 한계가 있음
  - 문제 발생시 순단 가능성
  - 각 서버는 직접 파일 시스템을 관리

- NAS(Network Attached Storage)

  ![스크린샷 2022-12-01 오후 11 22 03](https://user-images.githubusercontent.com/96982072/205081185-eb320ca3-b2d0-4436-970c-36fb34290d75.png)

  - 서버와 저장장치가 이더넷 등의 LAN 방식의 네트워크에 연결된 방식 
  - LAN은 TCP/IP 프로토콜 기반, 저장장치는 SCSI사용
    - 통신을 위해 중계역할의 파일서버 필요 
  - 확장성, 유연성이 뛰어남
  - 경제적, 설치 쉬움
  - 커넥션 증가 시 성능문제 발생 가능성

- SAN(Storage Area Network)

  ![스크린샷 2022-12-01 오후 11 22 12](https://user-images.githubusercontent.com/96982072/205081287-c39e647e-42cf-4a97-a87d-a7f3b3194fc9.png)

  - 서버와 저장장치를 광섬유 스위치로 연결한 고속 데이터 네트워크 
  - SAN 스위치를 구성한 뒤 서버와 스토리지를 연결
  - 확장성, 유연성, 가용성 우수

## DAS, NAS, SAN 비교
  
|  | DAS | NAS| SAN |
|---|:---:|:---:|:---:|
|구성요소|서버, 저장장치|서버, 전용파일서버, 저장장치|서버, FC 스위치, 저장장치|
|접속장치|X|이더넷 스위치로 접속|광섬유 스위치로 접속|
|규모|소규모|중규모|대규모|
|파일시스템 공유|불가능|가능|불가능|
|주요 사용처| |파일 공유, 동시 접속이 필요한 경우|DB, 가상화 등 대규모의 처리가 필요할 경우|
|비용| |(상대적으로) 낮음|(상대적으로) 높음|

## 데이터 보호를 위한 별의별 아이디어 - RAID

- RAID(Redundant Array of Inexpensive Disks)
  - 반드시 백업이 필요한 경우
  - 스토리지에 물리적인 이상이 발생하더라도 사용자가 체감할 수 없어야 함
  - 용량 증설 검토 시 데이터의 손실 없이 증설 필요
  - 여러 개의 디스크를 배치하여 속도↑,안정성↑,효율성↑, 가용성↑

- 크게 미러링(mirroring), 스트라이핑(striping)으로 나눠짐

- RAID 0 (Stripe)
  
  ![스크린샷 2022-12-01 오후 11 28 37](https://user-images.githubusercontent.com/96982072/205081522-df34ea5b-f5dd-487c-be17-ca9dbd3e0fc8.png)

  - 두 개 이상의 디스크에 데이터를 무작위로 write
  - 장점
    - 데이터 사용시 I/O를 디스크 개수만큼 분할하여 쓰기 때문에 I/O 속도가 향상됨
  - 단점
    - 디스크 중 하나의 디스크라도 장애가 발생하면 복구가 어려움

- RAID 1 (Mirror)
  
  ![스크린샷 2022-12-01 오후 11 31 51](https://user-images.githubusercontent.com/96982072/205081529-769273d6-662c-4ed5-886a-f02720a616a5.png)

  - 디스크에 동일한 데이터를 중복 기록.적어도 동일한 디스크 두 개가 필요
  - 장점
    - 디스크 두 개가 한번에 오류가 발생하지 않는 이상 가용성이 높고, 복구가 쉬움 
  - 단점
    - 사용할 수 있는 디스크가 실제 디스크의 절반
  
- RAID 2 (Stripe)

  ![스크린샷 2022-12-01 오후 11 33 11](https://user-images.githubusercontent.com/96982072/205081531-cb7f8fc6-029d-4d90-832e-a337732ff7bd.png)

  - 에러 체크, 수정을 위한 해밍코드 사용. ECC(Error Correction Code)를 별도의 드라이브에 저장
  - 단점: ECC가 저장된 드라이브 손상의 경우 복구가 어려움  
  (현재는 거의 사용되지 않음. deprecated)

- RAID 3,4 (Stripe)

  ![스크린샷 2022-12-01 오후 11 33 40](https://user-images.githubusercontent.com/96982072/205081730-f039b90f-bf41-43ec-bb2c-349445cd3399.png)

  - 에러 체크 및 수정을 위해 패리티 정보를 별도의 디스크에 저장
  - RAID 3은 데이터를 바이트 단위로 나누어 분산, RAID4는 블록단위로 나눔 
  - 장점
    - 성능 보완.디스크 용량을 온전히 사용할 수 있음
  - 단점
    - 별도의 패리티 디스크의 손상시 리스크 존재

- RAID 5 (Stripe)

  ![스크린샷 2022-12-01 오후 11 34 40](https://user-images.githubusercontent.com/96982072/205081735-224cb442-053d-4d94-8ede-7284b2e27ed2.png)

  - 패리티 정보 자체를 stripe로 구성된 디스크 내에서 처리하게 만듬
  - 장점
    - 하나의 디스크가 손상되더라도 남은 디스크들로 데이터 복구가 가능

- RAID 6 (Stripe)

  ![스크린샷 2022-12-01 오후 11 34 55](https://user-images.githubusercontent.com/96982072/205081739-8a15f6fd-7b83-4da6-b3fb-82465327c25b.png)

  - 기본적인 개념은 RAID5와 동일. 2차 패리티 정보를 넣어 2개의 디스크에 문제가 생겨도 복구가 가능
  - 장점
    - 안정성이 굉장히 높음
  - 단점
    - 약간의 데이터 공간 사용이 추가적으로 발생
