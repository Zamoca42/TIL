## 위상정렬 Topological Sorting

- 비순환 방향그래프(DAG: Directed Acyclic Graph)를 순서대로 출력

- 순서가 있는 task에서 순서를 찾아주는 알고리즘

  ![스크린샷 2022-11-17 오전 12 26 50](https://user-images.githubusercontent.com/96982072/202222209-a335e290-8660-46e9-b796-217ee99b2dcf.png)

- Queue로 구현하기
  - 진입차수(indegree) : 한 노드에 들어오는 다른 간선의 수

    1. 모든 vertex의 indegree 수를 센다
    2. 큐에 indegree가 0인 vertex 삽입
    3. 큐에서 vertex를 꺼내 연결된 edge 제거
    4. 3번으로 인해 indegree가 0이된 vertex를 큐에 삽입
    5. 큐가 빌 때까지 3-4 반복
