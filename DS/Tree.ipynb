# 트리 구조

- 한 노드가 여러 노드를 가리킬 수 있음
- 비선형적 자료구조
- 그래프라는 자료구조의 일종

- 데이터 구조의 계층적인 속성을 표현
	- 대표적으로 파일-폴더 구조

![img](https://user-images.githubusercontent.com/96982072/201464909-1c434ef5-237c-4f35-9ee3-bdc278f967fe.png)

- 트리(Tree) : Node와 Branch를 이용하여, 사이클을 이루지 않도록 구성한 데이터 구조
- 노드(Node) : 트리에서 데이터를 저장하는 기본 요소로, Branch 정보 또한 포함
- 간선(Edge) : 노드와 노드를 연결하는 연결선
- Root Node : 트리 맨 위의 노드
- Leaf Node : Child Node가 하나도 없는 노드
- Depth : 트리에서 Node가 가질 수 있는 최대 Level
- Parent node : 간선으로 연결된 노드 중 상위 노드
- Child node : 간선으로 연결된 노드 중 하위 노드

# 트리 순회

-  Pre-order 
  - **A**–B–D–H–E–C–F–I–J–G–K
-  In-order 
  - D–H–B–E–**A**–I–F–J–C–G–K
-  Post-order 
  - H–D–E–B–I–J–F–K–G–C–**A**

- 탐색 종류에 따라 루트 노드(A) path가 다름

# 이진 탐색 트리 (Binary Search Tree - BST)

- 트리 구조 자체로는 데이터의 특성에 아무런 제약이 없음

  - 노드의 왼쪽 서브 트리에는 루트 노드보다 작은 값
  - 노드의 오른쪽 서브 트리에는 루트 노드보다 큰 값
  - 서브 트리는 다시 이진 탐색 트리
  - 중복된 값은 없음  
![스크린샷 2022-11-12 오후 5 56 51](https://user-images.githubusercontent.com/96982072/201471445-22c1d0d1-6d27-425f-bb39-82fb4dcaedfd.png)


```python
class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class BinarySearchTree:
    def __init__(self):
        self.root = None
        self._size = 0
    # 중복된 데이터는 삽입하지 않음
    # 추가된 노드는 트리의 leaf에 삽입
    def insert_node(self, node, val):
        if node is None:
            return Node(val)
        if val < node.val: # val이 노드의 값보다 작은 경우
            node.left = self.insert_node(node.left, val) # 왼쪽에 삽입
        elif val > node.val: # val이 노드의 값보다 큰 경우
            node.right = self.insert_node(node.right, val) # 오른쪽에 삽입
        return node

    def size(self):
        return self._size

    def insert(self, val):
        self.root = self.insert_node(self.root, val)
        self._size += 1

    def contains(self, val):
        return self.contains_node(self.root, val)

    def contains_node(self, node, val):
        if node is None: # 찾고자 하는 값이 없는 경우 false
            return False
        if val == node.val: # 트리에 값이 존재
            return True
        if val < node.val: # 찾는 값이 노드의 값보다 작을 떄
            return self.contains_node(node.left, val) # 왼쪽 트리에 리턴
        return self.contains_node(node.right, val)
    
    # 삭제
    def delete(self, val):
        return self.delete_node(self.root, val)

    def delete_node(self, node, val):
        if node is None:
            return None

        if val < node.val:
            node.left = self.delete_node(node.left, val)
        elif val > node.val:
            node.right = self.delete_node(node.right, val)
        else:
            self._size -= 1
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            node.val = self.min_mode(node.right)
            node.right = self.delete_node(node.right, node.val)
        return node

    def min_node(node):
        min_val = node.val
        while node.left is not None:
            min_val = node.left.val
            node = node.left
        return min_val

    def preorder(self):
        ret = []

        def visit(root):
            nonlocal ret
            if root is None:
                return ret
            # 전위 탐색
            #1. 루트 노드 방문
            #2. 왼쪽 서브트리
            #3. 오른쪽 서브트리
            ret.append(root.val)
            visit(root.left)
            visit(root.right)

        visit(self.root)
        return ret

    def inorder(self):
        ret = []

        def visit(root):
            nonlocal ret
            if root is None:
                return ret
            # 중위 탐색
            #1. 왼쪽 서브트리
            #2. 루트 노드 방문
            #3. 오른쪽 서브트리
            visit(root.left)
            ret.append(root.val)
            visit(root.right)

        visit(self.root)
        return ret

    def postorder(self):
        ret = []

        def visit(root):
            nonlocal ret
            if root is None:
                return ret
            # 후위 탐색
            #1. 왼쪽 서브트리
            #2. 오른쪽 서브트리
            #3. 루트 노드 방문

            visit(root.left)
            visit(root.right)
            ret.append(root.val)

        visit(self.root)
        return ret


if __name__ == "__main__":
    btree = BinarySearchTree()
    for i in [3, 1, 2, 3, 8, 7]:
        print(f"btree.insert({i})")
        btree.insert(i)
        print(f"btree.size(): {btree.size()}")

    print("==================================")
    for i in [3, 1, 2, 3, 8, 7]:
        print(f"btree.contains({i}): {btree.contains(i)}")

    print("===================================")
    print(f"btree.preorder(): {btree.preorder()}")
    print(f"btree.inorder(): {btree.inorder()}")
    print(f"btree.postorder(): {btree.postorder()}")
    print("=====================================")

    print("btree.delete(2)")
    btree.delete(2)
    print("btree.delete(8)")
    btree.delete(8)
    print(f"btree.size(): {btree.size()}")

    print("==================================")
    for i in [3, 1, 2, 3, 8, 7]:
        print(f"btree.contains({i}): {btree.contains(i)}")

    print("===================================")
    print(f"btree.preorder(): {btree.preorder()}")
    print(f"btree.inorder(): {btree.inorder()}")
    print(f"btree.postorder(): {btree.postorder()}")
    print("=====================================")
```
